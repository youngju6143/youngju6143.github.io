### 도미 데이터 준비하기

### 빙어 데이터 준비하기

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

## 첫 번째 머신러닝 프로그램

```python
length = bream_length+smelt_length
weight = bream_weight+smelt_weight
```

```python
# Code Problem 1
# 데이터를 사이킷런에서 사용할 수 있는 형태로 만들기
# zip() 함수와 리스트 내포 구문 사용하여 2차원 리스트로 만들기
fish_data = [[l, w] for l, w in zip(length, weight)]
print(fish_data)
```

```python
# Code Problem 2
# 정답 데이터 만들기
# 도미 데이터(35개)를 찾는 대상으로 설정
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
# Code Problem 3
# kn 모델 학습시키기
# 사이킷런에서 훈련하기 위해 사용하는 메서드에 데이터를 매개변수로 전달하기
kn.fit(fish_data, fish_target)
```

```python
# Code Problem 4
# kn 모델 성능 평가하기(정확도)
# 사이킷런에서 모델을 평가하기 위해 사용하는 메서드에 데이터를 매개변수로 전달하기
kn.score(fish_data, fish_target)
```

```python
# Code Problem 5
# 새로운 데이터의 정답 예측하기
# 새로운 데이터의 정답을 예측하기 위해 사용하는 메소드에 삼각형 포인트의 데이터 넣기
# 리스트의 리스트를 전달해야 함
kn.predict([[30, 600]])
```

반명 다른 머신러닝 모델에서의 훈련은 모델의 파라미터를 학습시킨다고 할 수 있습니다. 데이터셋의 패턴을 학습하여 모델의 가중치를 조정합니다.

답변: 이런 현상이 나타나는 이유는 샘플링 편향 때문입니다. KNN은 가장 가까운 이웃들을 근거로 정답을 예측합니다.

따라서 모델의 정확도는 도미의 수/전체 개수 가 될 수 있습니다.

# 데이터 전처리

## 넘파이로 데이터 준비하기

```python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
import numpy as np
```

```python
# Code Problem 11
# 넘파이 함수를 이용하여 리스트처럼 한줄로 길게 출력되지 않고
# 행과 열을 맞추어 가지런히 정리된 모습으로 출력하는 코드

fish_data = np.column_stack((fish_length, fish_weight))
```

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
```

```python
print(fish_target)
```

## 사이킷런으로 훈련 세트와 테스트 세트 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state=42)
```

```python
print(train_input.shape, test_input.shape)
```

```python
print(train_target.shape, test_target.shape)
```

```python
print(test_target)
```

```python
# Code Problem 12
# 현재 test_target의 비율이 편향이 되어 있는 상태이다.
# 클래스 비율에 맞게 데이터를 나눠주는 매개변수를 활용해서 다시 정의하시오.

train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify=fish_target, random_state=42)
```

```python
print(test_target)
```

## 수상한 도미 한마리

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
print(kn.predict([[25, 150]]))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
distances, indexes = kn.kneighbors([[25, 150]])
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
